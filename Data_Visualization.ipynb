```python
df = pd.read_csv('/content/drive/MyDrive/BS/data_visualization/sharma-kohli.csv')
```

```python
df.head()
```

```python
plt.plot(df['index'], df['RG Sharma'], color="red",  linestyle="dashdot", linewidth=3, label='Sharma', marker='D')
plt.plot(df['index'], df['V Kohli'],color="green", linestyle="dashed", linewidth=2, label='Kohli', marker='+')

plt.xlabel("Season")
plt.ylabel("Run")
plt.title("Sharma vs Kohli")
plt.legend()
plt.show()
```

# Scatter Plot

```python
x = np.linspace(-10, 10, 50)
```

```python
y = 3 * x + 5 + np.random.randint(0, 100, 50)
```

```python
plt.scatter(x, y, marker=".", label="Points", color="red")
plt.legend()
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
df_batter = pd.read_csv('/content/drive/MyDrive/BS/data_visualization/batter.csv')
df_batter = df_batter.head(20)
```

```python
plt.scatter(df_batter["avg"], df_batter["strike_rate"], marker="^")
plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.title("Average vs Strike Rate")
plt.show()
```
