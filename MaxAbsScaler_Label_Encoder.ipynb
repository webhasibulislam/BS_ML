# MaxAbsScaler

```python
df.describe()
```

```python
from sklearn.preprocessing import MaxAbsScaler
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:, [2, 4]]
y = df.iloc[:, 5]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=50)
```

```python
scale_maxabs = MaxAbsScaler().fit(x_train)
```

```python
print(scale_maxabs)
```

```python
scaled_x_train = scale_maxabs.transform(x_train)
```

```python
print(scaled_x_train)
```

```python
scale_x_test = MaxAbsScaler().fit(x_test)
```

```python
x_test.head()
```

```python
scaled_x_test = scale_x_test.transform(x_test)
```

```python
print(scaled_x_test)
```
