# RobustScaler

```python
df_rs = pd.read_csv('/content/drive/MyDrive/BS/autompg.csv')
```

```python
df_rs.head()
```

```python
df_rs.tail()
```

```python
df_rs.dtypes
```

```python
x_rs = df_rs.iloc[:, [2, 4]]
y_rs = df_rs.iloc[:, 5]
```

```python
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
```

```python
x_train_rs, x_test_rs, y_train_rs, y_test_rs = train_test_split(x_rs, y_rs, random_state = 40)
```

```python
x_train_rs.head()
```

```python
x_test_rs.head()
```

```python
scale_rs = MinMaxScaler().fit(x_train_rs)
```

```python
print(scale_rs)
```

```python
scale_rs.data_min_
```

```python
scale_rs.data_max_
```

```python
scaled_rs_x_train = scale_rs.transform(x_train_rs)
```

```python
print(scaled_rs_x_train)
```

**RobustScale of Test Data**

```python
scale_rs_x_test = RobustScaler().fit(x_test_rs)
```

```python
scaled_rs_x_test = scale_rs_x_test.transform(x_test_rs)
```

```python
print(scaled_rs_x_test)
```
